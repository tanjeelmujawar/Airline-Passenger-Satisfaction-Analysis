```python
rating_cols = [
    'Departure and Arrival Time Convenience',
    'Ease of Online Booking',
    'Check-in Service',
    'Online Boarding',
    'Gate Location',
    'On-board Service',
    'Seat Comfort',
    'Leg Room Service',
    'Cleanliness',
    'Food and Drink',
    'In-flight Service',
    'In-flight Wifi Service',
    'In-flight Entertainment',
    'Baggage Handling'
]


df[rating_cols] = df[rating_cols].replace(0, np.nan)
```

```python
df['Arrival Delay'].fillna(df['Arrival Delay'].median(), inplace=True)
```

```python

df['Age Group'] = pd.cut(
    df['Age'],
    bins = [0,24,55,100],
    labels=['Young', 'Adult', 'Senior']
)
```

```python
df.head()
```

```python

df['Delay Flag'] = np.where(df['Arrival Delay'] > 15, 'Delayed', 'On Time')
```

```python

df['Service Quality Index'] = df [[ 'Online Boarding', 'Seat Comfort', 'Cleanliness', 'In-flight Service']].mean(axis=1)
```
